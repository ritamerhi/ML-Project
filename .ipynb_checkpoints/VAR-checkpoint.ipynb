## VAR

```python
import numpy as np
import pandas as pd
import yfinance as yf
from statsmodels.tsa.api import VAR
from sklearn.preprocessing import StandardScaler
import pickle
import datetime
from datetime import datetime, timedelta

###############################################
# 1. Data Downloading + Preprocessing
###############################################

def load_prices(start_date, end_date):
    gold = yf.download("GC=F", start=start_date, end=end_date, interval="30m", auto_adjust=False)
    silver = yf.download("SI=F", start=start_date, end=end_date, interval="30m", auto_adjust=False)
    cad = yf.download("CADUSD=X", start=start_date, end=end_date, interval="30m", auto_adjust=False)

    gold = gold[["Close"]].rename(columns={"Close": "Gold"})
    silver = silver[["Close"]].rename(columns={"Close": "Silver"})
    cad = cad[["Close"]].rename(columns={"Close": "CAD"})

    all_prices = gold.join([silver, cad], how="outer")
    all_prices = all_prices.ffill().bfill()

    return all_prices


###############################################
# 2. Train VAR Model
###############################################

def train_var(prices, maxlags=48, model_path="var_model.pkl"):
    scaler = StandardScaler()
    scaled = scaler.fit_transform(prices)
    scaled_df = pd.DataFrame(scaled, index=prices.index, columns=prices.columns)

    model = VAR(scaled_df)
    results = model.fit(maxlags)

    # save model + scaler
    with open(model_path, "wb") as f:
        pickle.dump((results, scaler, prices.columns.tolist()), f)

    print("VAR model trained & saved!")
    return results, scaler


###############################################
# 3. Load VAR Model
###############################################

def load_var_model(model_path="var_model.pkl"):
    with open(model_path, "rb") as f:
        results, scaler, columns = pickle.load(f)
    return results, scaler, columns


###############################################
# 4. Forecasting Function
###############################################

def forecast_interval(start_date, end_date, model_path="var_model.pkl"):
    """
    Predicts 30-min prices between start_date and end_date.
    """
    results, scaler, cols = load_var_model(model_path)

    # number of 30-min steps
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)
    delta = end - start
    steps = int(delta.total_seconds() // (30 * 60))
    if steps <= 0:
        raise ValueError("end_date must be after start_date")

    # last observed training data point
    last_obs = results.model.endog[-results.k_ar:]


    # forecast (scaled)
    forecast_scaled = results.forecast(last_obs, steps=steps)

    # inverse scaling
    forecast = scaler.inverse_transform(forecast_scaled)

    # create timestamp index for the predictions
    times = pd.date_range(start=start, periods=steps, freq="30min")

    forecast_df = pd.DataFrame(forecast, index=times, columns=cols)
    return forecast_df


###############################################
# 5. Example: Train the Model
###############################################

# choose some historical period to train on
train_end   = datetime.now()
train_start = train_end - timedelta(days=60)

prices = load_prices(train_start, train_end)
results, scaler = train_var(prices, maxlags=48)


###############################################
# 6. Example: Predict Any Future Interval
###############################################

# user input interval
pred_start = "2025-10-01"
pred_end   = "2025-10-02"

forecast_df = forecast_interval(pred_start, pred_end)
print(forecast_df)
```

```python
import numpy as np

def monte_carlo_var(results, scaler, cols, steps, sims=50):
    """
    Generates Monte Carlo simulated price paths based on a trained VAR model.
    
    results : VARResults (trained model)
    scaler  : scaler used for training
    cols    : ['Gold', 'Silver', 'CAD']
    steps   : number of 30min steps to forecast
    sims    : number of simulated paths to generate
    
    Returns: sims x steps x 3 numpy array (in original price scale)
    """
    
    # VAR residual covariance  used for noise
    cov = results.sigma_u
    
    # Standard normal innovations for each simulation and step
    shocks = np.random.multivariate_normal(mean=np.zeros(len(cols)), cov=cov, size=(sims, steps))
    
    # last observed window required for VAR
    k_ar = results.k_ar
    history = results.model.endog[-k_ar:]
    
    all_paths = []
    
    for s in range(sims):
        obs = history.copy()
        path_scaled = []
        
        for t in range(steps):
            # deterministic VAR forecast
            pred = results.forecast(obs, 1)[0]
            
            # add noise shock
            pred_noisy = pred + shocks[s, t]
            
            path_scaled.append(pred_noisy)
            
            # update rolling window
            obs = np.vstack([obs[1:], pred_noisy])
        
        # inverse scaling  actual prices
        path = scaler.inverse_transform(np.array(path_scaled))
        all_paths.append(path)
    
    return np.array(all_paths)
```

```python
def forecast_interval_with_mc(start_date, end_date, sims=50, model_path="var_model.pkl"):
    results, scaler, cols = load_var_model(model_path)

    start = pd.to_datetime(start_date)
    end   = pd.to_datetime(end_date)
    steps = int((end - start).total_seconds() // (30 * 60))

    if steps <= 0:
        raise ValueError("end_date must be after start_date")

    # Deterministic VAR forecast
    last_obs = results.model.endog[-results.k_ar:]
    forecast_scaled = results.forecast(last_obs, steps=steps)
    forecast = scaler.inverse_transform(forecast_scaled)

    times = pd.date_range(start=start, periods=steps, freq="30min")
    forecast_df = pd.DataFrame(forecast, index=times, columns=cols)

    # Monte Carlo simulated paths
    mc_paths = monte_carlo_var(results, scaler, cols, steps, sims)

    return forecast_df, mc_paths, times
```

```python
forecast_df, mc_paths, times = forecast_interval_with_mc("2025-10-01", "2025-10-02", sims=50)
```

```python
import matplotlib.pyplot as plt

def plot_mc(forecast_df, mc_paths, times, asset_index=0, asset_name="Gold"):
    plt.figure(figsize=(12,6))
    
    # plot MC paths
    for i in range(mc_paths.shape[0]):
        plt.plot(times, mc_paths[i, :, asset_index], color="gray", alpha=0.2)
    
    # plot deterministic VAR forecast
    plt.plot(times, forecast_df.iloc[:, asset_index], color="red", linewidth=2, label="VAR forecast")
    
    plt.title(f"Monte Carlo VAR Simulation for {asset_name}")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_mc(forecast_df, mc_paths, times, asset_index=0, asset_name="Gold")
plot_mc(forecast_df, mc_paths, times, asset_index=1, asset_name="Silver")
plot_mc(forecast_df, mc_paths, times, asset_index=2, asset_name="CAD")
```
