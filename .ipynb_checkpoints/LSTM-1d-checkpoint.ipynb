## feature engineering

```python
# adding features that make more sense than raw data
all_prices["Gold_ret"] = all_prices["Gold"].pct_change()
all_prices["Silver_ret"] = all_prices["Silver"].pct_change()
all_prices["CAD_ret"] = all_prices["CAD"].pct_change()

all_prices["Gold_ma"] = all_prices["Gold"].rolling(5).mean()
all_prices["Silver_ma"] = all_prices["Silver"].rolling(5).mean()
all_prices["CAD_ma"] = all_prices["CAD"].rolling(5).mean()

all_prices = all_prices.dropna()
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(all_prices)

scaled_prices = pd.DataFrame(scaled, index=all_prices.index, columns=all_prices.columns)
```

```python
# gold only
gold = gold[["Close"]].rename(columns={"Close": "Gold"})
scaler = MinMaxScaler()
scaled = scaler.fit_transform(gold)

scaled_prices = pd.DataFrame(scaled, index=gold.index, columns=gold.columns)
```

```python
gold.info()
```

```python
def create_sequences(data, seq_len=20):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i : i + seq_len])
        y.append(data[i + seq_len, 0]) # gold price is column 0
    return np.array(X), np.array(y)

SEQ_LEN = 20
X, y = create_sequences(scaled_prices.values, SEQ_LEN)

X.shape, y.shape
```

```python
# split = int(len(X) * 0.8)

# X_train, X_test = X[:split], X[split:]
# y_train, y_test = y[:split], y[split:]

X_train = X[:-1]
y_train = y[:-1]

X_test = X[-1:]          # last window
y_test = y[-1:]          # today's gold price (the next real value)
```

```python
model = Sequential([
    layers.Input(shape=(SEQ_LEN, X.shape[2])),
    LSTM(64, return_sequences=True),
    Dropout(0.2),

    LSTM(32),
    Dropout(0.2),

    Dense(16, activation='relu'),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    epochs=25,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

```python
# # Predict
y_pred_scaled = model.predict(X_test)

# # Inverse transform
# full_scaled = scaled_prices.values

# y_test_inv = []
# y_pred_inv = []

# for i in range(len(y_test)):
#     row = full_scaled[-len(y_test) + i].copy()
#     row[0] = y_test[i]
#     y_test_inv.append(scaler.inverse_transform([row])[0][0])

#     row2 = full_scaled[-len(y_test) + i].copy()
#     row2[0] = y_pred_scaled[i]
#     y_pred_inv.append(scaler.inverse_transform([row2])[0][0])

# y_test_inv = np.array(y_test_inv)
# y_pred_inv = np.array(y_pred_inv)


# Reconstruct full row to inverse transform properly
last_row = scaled_prices.iloc[-1].copy().values

# TRUE: today's gold price
true_row = last_row.copy()
true_row[0] = y_test[0]
y_test_inv = scaler.inverse_transform([true_row])[0][0]

# PREDICTED: today's gold price
pred_row = last_row.copy()
pred_row[0] = y_pred_scaled[0]
y_pred_inv = scaler.inverse_transform([pred_row])[0][0]

print("True Today's Gold Price:", y_test_inv)
print("Predicted Today's Gold Price:", y_pred_inv)
```

# LSTM

```python
window = 20

def create_sequences(data, sequence):
    X = [] 
    y = []
    for i in range(len(data) - sequence):
        X.append(data[i : i + sequence])
        y.append(data[i + sequence, :3]) # predict gold, silver, cad
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_prices.values, window)
```

```python
split = int(len(X) * 0.8)
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]
```

```python
print(y_train.shape)
print(y_val.shape)
```

```python
model = models.Sequential([
    layers.Input(shape=(window, scaled_prices.shape[1])),
    layers.LSTM(64, return_sequences=True),
    layers.LSTM(32),
    layers.Dense(3)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=32
)
```

```python
# predicting one step ahead then appending that prediction into the sequence
def walk_forward_predict(model, X, df, seq_len=20, steps=20):
    preds = []
    current_seq = X[-1]  
    F = df.shape[1] 

    for _ in range(steps):
        pred_scaled = model.predict(current_seq[np.newaxis, :, :])[0]  
        new_row = np.zeros(F)
        new_row[0:3] = pred_scaled 
        new_row[3:] = current_seq[-1, 3:]
        
        # shifting window forward
        current_seq = np.vstack([current_seq[1:], new_row])

        preds.append(pred_scaled)

    return np.array(preds)
```

```python
scaled_future_preds = walk_forward_predict(
    model,
    X_val,
    df=scaled_prices,
    seq_len=window,
    steps=20
)
```

```python
steps = scaled_future_preds.shape[0]
extra = scaled_prices.shape[1] - 3

# padding to full dimension for inverse scaling
padded = np.hstack([
    scaled_future_preds,
    np.zeros((steps, extra))
])

future_full = scaler.inverse_transform(padded)

future_gold  = future_full[:, 0]
future_silver = future_full[:, 1]
future_cad    = future_full[:, 2]
```

```python
actual_scaled = y_val[:20]

extra = scaled_prices.shape[1] - 3
actual_padded = np.hstack([
    actual_scaled,
    np.zeros((actual_scaled.shape[0], extra))
])

actual_full = scaler.inverse_transform(actual_padded)

actual_gold  = actual_full[:, 0]
actual_silver = actual_full[:, 1]
actual_cad    = actual_full[:, 2]
```

```python
# persistence baseline
baseline_gold  = np.roll(actual_gold, 1)
baseline_silver = np.roll(actual_silver, 1)
baseline_cad    = np.roll(actual_cad, 1)

baseline_gold[0] = actual_gold[0]
baseline_silver[0] = actual_silver[0]
baseline_cad[0] = actual_cad[0]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# gold
plt.figure(figsize=(14,6))
plt.plot(actual_gold, color="blue", label="Actual Gold")
plt.plot(future_gold, color="orange", label="Predicted Gold")
plt.plot(baseline_gold, color="magenta", label="Baseline Gold")
plt.title("Gold Price Prediction (30m)")
plt.xlabel("Time (Validation Steps)")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# silver
plt.figure(figsize=(14,6))
plt.plot(actual_silver, color="blue", label="Actual Silver")
plt.plot(future_silver, color="orange", label="Predicted Silver")
plt.plot(baseline_silver, color="magenta", label="Baseline Silver")
plt.title("Silver Price Prediction (30m)")
plt.xlabel("Time (Validation Steps)")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# CAD
plt.figure(figsize=(14,6))
plt.plot(actual_cad, color="blue", label="Actual CAD/USD")
plt.plot(future_cad, color="orange", label="Predicted CAD/USD")
plt.plot(baseline_cad, color="magenta", label="Baseline CAD/USD")
plt.title("CAD/USD Prediction (30m)")
plt.xlabel("Time (Validation Steps)")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```
